## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

import matplotlib.pyplot as plt
import seaborn as sns

#define data
Query="""
SELECT DatabaseDesireNextYear, count(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc limit 5
"""

DatabaseDesireNextYear= pd.read_sql_query(Query, conn)
x = DatabaseDesireNextYear['count']

labels = DatabaseDesireNextYear['DatabaseDesireNextYear']

#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]

#create pie chart
plt.pie(x, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
#In the list of most popular languages respondents wish to learn next year, what is the rank of Python?
Query="""
SELECT LanguageDesireNextYear, count(*) as count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count desc 
"""
df = pd.read_sql_query(Query, conn)

# sort the dataframe by count in descending order
df = df.sort_values(by='count', ascending=False)

# reset the index
df = df.reset_index(drop=True)

# add a rank column
df['rank'] = df.index + 1

# find the row for Python and print its rank
python_row = df.loc[df['LanguageDesireNextYear'] == 'Python']
print(f"The rank of Python is {python_row['rank'].item()}")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query = """
SELECT Avg(WorkWeekHrs), Avg(CodeRevHrs), Age 
FROM master 
where Age between 30 and 35
group by Age
"""

df = pd.read_sql_query(Query, conn)

WorkWeekHrs= df['Avg(WorkWeekHrs)']
CodeRevHrs= df['Avg(CodeRevHrs)']
Age= df['Age']

fig, ax = plt.subplots()

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.show()
```

### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.
